```python
#기대수명 평균이 가장 낮은 국가는?
grp_df = df.groupby('country')['lifeExp'].mean().idxmin()
cou_grp_df = df.groupby('country')['lifeExp'].mean().min()

print(f'기대수명이 가장 낮은 나라는 {grp_df}이며, 평균 기대 수명은 {int(cou_grp_df)}살 입니다')
```

```python
import matplotlib.pyplot as pl
```

```python
# 기대수명이 가장낮은 대륙은?
idx_con_grp_df = df.groupby('continent')['lifeExp'].mean().idxmin()
con_grp_df = df.groupby('continent')['lifeExp'].mean().min()
print(f'기대수명이 가장 낮은 대륙은 {idx_con_grp_df}이며, 평균 기대 수명은 {int(con_grp_df)}살 입니다')
mean_x = con_grp_df = df.groupby('continent')['lifeExp'].mean()
mean_x.plot()

type(mean_x)
```

```python
# 기대수명이 가장낮은 대륙은?
idx_con_grp_df = {df.groupby('continent')['lifeExp'].mean().idxmin(): int(df.groupby('continent')['lifeExp'].mean().min())}
(idx_con_grp_df.values()).plot


```
