```python
import matplotlib.pyplot as plt
plt.rcParamsDefault['font.family'] = 'Noto Sans KR'
plt.rcParams['font.family'] = 'Noto Sans KR'

# 간단한 예제
x = [1,2,3,4,5]
y = [2,4,6,8,10]    

plt.plot(x,y) # 선그래프
plt.title('선그래프')
plt.xlabel('행 범례')
plt.ylabel('열 범례')
plt.show()
```

```python
ca = ['지훈', '해림', '라희']
va = [31,25,2]

plt.bar(ca,va)
plt.title('이름과나이')
plt.xlabel('이름')
plt.ylabel('나이')
plt.show()
```

## (3) 히스토그램

분포를 이해할 때 자주 사용


```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Noto Sans KR'

data = np.random.randn(1000) # 평균 0, 표준편차 1인 데이터 1000개 생성


plt.hist(data, bins=30)
plt.title('히스토그램 예시')
plt.xlabel('분포')
plt.ylabel('행높이')
plt.show()
```

## (4) 박스플롯
데이터의 이상치나 분포의 범위 파악

```python
data = np.random.randn(1000)

plt.boxplot(data)
plt.title('box')
plt.xlabel('행')
plt.ylabel('열')
plt.show()
```
