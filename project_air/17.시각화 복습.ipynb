```python
import matplotlib.pyplot as plt
plt.rcParamsDefault['font.family'] = 'Noto Sans KR'
plt.rcParams['font.family'] = 'Noto Sans KR'

# 간단한 예제
x = [1,2,3,4,5]
y = [2,4,6,8,10]    

plt.plot(x,y) # 선그래프
plt.title('선그래프')
plt.xlabel('행 범례')
plt.ylabel('열 범례')
plt.show()
```

```python
ca = ['지훈', '해림', '라희']
va = [31,25,2]

plt.bar(ca,va)
plt.title('이름과나이')
plt.xlabel('이름')
plt.ylabel('나이')
plt.show()
```
